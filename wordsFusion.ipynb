```python
from arabert.preprocess import ArabertPreprocessor
model_name="bert-base-arabertv2"
arabert_prep = ArabertPreprocessor(model_name=model_name)

text = "قَدّ خَرَجَتْ سَالِي."
arabert_prep.preprocess(text)
```

```python
from transformers import AutoTokenizer, AutoModel
import torch
import torch.nn.functional as F

model_name = 'aubmindlab/bert-base-arabert'
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
model= model.cuda()

sentence = "Closer to the axis or confrontation of an organ or organism"
word = "nearest to or facing toward the axis of an organ or organism"

tokens = tokenizer.tokenize(sentence)
word_tokens = tokenizer.tokenize(word)

token_ids = tokenizer.convert_tokens_to_ids(tokens)
word_token_ids = tokenizer.convert_tokens_to_ids(word_tokens)

with torch.no_grad():
    inputs = torch.tensor([token_ids]).cuda()
    outputs = model(inputs)
    sentence_embeddings = outputs.last_hidden_state.mean(dim=1).squeeze()

    word_inputs = torch.tensor([word_token_ids]).cuda()
    word_outputs = model(word_inputs)
    word_embedding = word_outputs.last_hidden_state.mean(dim=1).squeeze()

# Calculate cosine similarity
cosine_sim = F.cosine_similarity(sentence_embeddings, word_embedding, dim=0).item()


print(f"Similarity between the sentence and word: {cosine_sim}")
```

```python
sentence = "Closest to or facing the axis of the organ or organism"
word = "nearest to or facing toward the axis of an organ or organism"

tokens = tokenizer.tokenize(sentence)
word_tokens = tokenizer.tokenize(word)
```

```python
def calculate_cosine_similarity_arabert(tokens, word_tokens , model , tokenizer):
    token_ids = tokenizer.convert_tokens_to_ids(tokens)
    word_token_ids = tokenizer.convert_tokens_to_ids(word_tokens)

    with torch.no_grad():
        inputs = torch.tensor([token_ids]).cuda()
        outputs = model(inputs)
        sentence_embeddings = outputs.last_hidden_state.mean(dim=1).squeeze()

        word_inputs = torch.tensor([word_token_ids]).cuda()
        word_outputs = model(word_inputs)
        word_embedding = word_outputs.last_hidden_state.mean(dim=1).squeeze()

    # Calculate cosine similarity
    cosine_sim = F.cosine_similarity(sentence_embeddings, word_embedding, dim=0).item()
    
    return cosine_sim
```

```python
def get_tokens(sentence,tokenizer):
  return tokenizer.tokenize(sentence)
```

```python
final = []
counter =0
for (deff , words) in zip(words_array , words_tot):
  if(counter>10789):
    definition = deff[1]
    tokens = get_tokens(arabert_prep.preprocess(definition),tokenizer)
    maxSim =0
    second = 0
    third = 0
    idx_max = -10
    idx_2=-10
    idx_3=-10
    for idx,element in enumerate(words):
      if(element !='') and (element !=' '):
        word_tokens = get_tokens(arabert_prep.preprocess(element),tokenizer)
        sim = calculate_cosine_similarity_arabert(tokens,word_tokens,model,tokenizer)
        if(sim > maxSim):
          maxSim = sim
          idx_max = idx
        elif( sim > second):
          second = sim
          idx_2 = idx
        elif(sim > third):
          third = sim
          idx_3 = idx
    step=[]
    if(idx_max != -10):
      step.append(maxSim)
      step.append(words[idx_max])
      if(idx_2 != -10):
        step.append(second)
        step.append(words[idx_2])
        if(idx_3 != -10):
          step.append(third)
          step.append(words[idx_3])
      final.append(step)
    else :
      final.append(words)
  print(counter)
  counter+=1
    
```
