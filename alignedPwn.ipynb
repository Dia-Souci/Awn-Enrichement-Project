```python
import xml.etree.ElementTree as ET

def create_xml_file(ids, names, definitions, examples):
    root = ET.Element("root")  # Create the root element

    for i in range(len(ids)):
        item = ET.SubElement(root, "LexicalEntry",id=str(ids[i]) )  # Create the item element

        forms = ET.SubElement(item, "WrittenForm")
        for element in  names[i] :
          name_element = ET.SubElement(forms, "Form")
          name_element.text = element

        def_element = ET.SubElement(item, "definition")  # Create the def element
        def_element.text = definitions[i]  # Set the def value

        examples_element = ET.SubElement(item, "Examples")

        for element in examples[i] :
          example_element = ET.SubElement(examples_element, "example")  # Create the example element
          example_element.text = element # Set the example value

    tree = ET.ElementTree(root)  # Create the XML tree
    tree.write("/content/drive/MyDrive/Final folder/pwnFilesAlined/smallResultsForValidation.xml", encoding="utf-8", xml_declaration=True)  # Save the tree to a file

# Sample data
create_xml_file(ali_ids[:200], ali_words, ali_def, ali_ex)
```

```python
import xml.dom.minidom as minidom

def create_xml_file(ids, names, definitions, examples):
    doc = minidom.Document()  # Create a new XML document

    root = doc.createElement("root")  # Create the root element
    doc.appendChild(root)

    for i in range(len(ids)):
        item = doc.createElement("LexicalEntry")  # Create the item element
        item.setAttribute("id", str(ids[i]))  # Set the "id" attribute

        forms = doc.createElement("WrittenForms")
        item.appendChild(forms)

        for element in  names[i] :
          name_element = doc.createElement("Form")  # Create the name element
          name_element.appendChild(doc.createTextNode(element))  # Set the name value
          forms.appendChild(name_element)


        lis = ids[i].split('.')
        if(lis[0] != ''):
          englishName = lis[0]
        else :
          englishName = lis[1]
        Englishform = doc.createElement("Englishform")
        Englishform.appendChild(doc.createTextNode(englishName))
        item.appendChild(Englishform)


        def_element = doc.createElement("Definition")  # Create the def element
        def_element.appendChild(doc.createTextNode(definitions[i] ))  # Set the def value
        item.appendChild(def_element)

        engDef = wn.synset(ids[i]).definition()
        engdef_element = doc.createElement("EngDefinition")  # Create the def element
        engdef_element.appendChild(doc.createTextNode(engDef ))  # Set the def value
        item.appendChild(engdef_element)



        examples_element = doc.createElement("Examples")
        item.appendChild(examples_element)

        for element in examples[i] :
          example_element = doc.createElement("example")  # Create the example element
          example_element.appendChild(doc.createTextNode(element))  # Set the example value
          examples_element.appendChild(example_element)

        root.appendChild(item)

    # Pretty print the XML
    xml_str = doc.toprettyxml(indent="  ", encoding="utf-8")

    with open("/content/drive/MyDrive/Final folder/pwnFilesAlined/ResultsForValidation1000.xml", "wb") as file:
        file.write(xml_str)

# Generate XML file
create_xml_file(ali_ids, ali_words, ali_def, ali_ex)
```
