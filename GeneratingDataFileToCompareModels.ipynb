```python
import torch
from transformers import AdamW, AutoModelForSeq2SeqLM, AutoTokenizer
```

```python
model_repo = 'google/mt5-small'
model_path = '/content/drive/MyDrive/MT5/v2MT5_AR_EN.pt'
max_seq_len = 100
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_repo)
model = AutoModelForSeq2SeqLM.from_pretrained(model_repo)
model = model.cuda()
```

```python
LANG_TOKEN_MAPPING = {
    'en': '<en>',
    'ar': '<ar>'
}

special_tokens_dict = {'additional_special_tokens': list(LANG_TOKEN_MAPPING.values())}
tokenizer.add_special_tokens(special_tokens_dict)
model.resize_token_embeddings(len(tokenizer))
```

```python
model.load_state_dict(torch.load(model_path))
```

```python
def encode_input_str(text, target_lang, tokenizer, seq_len, lang_token_map=LANG_TOKEN_MAPPING):
  
  target_lang_token = lang_token_map[target_lang]

  input_ids = tokenizer.encode(
      text = target_lang_token + text,
      return_tensors = 'pt',
      padding = 'max_length',
      truncation = True,
      max_length = seq_len)

  return input_ids[0]

def handleUndecisive(tokens):
  if(tokens[-1] != 1):
    redun = tokens[-1]
    count =0
    for e in tokens :
      if (e==redun):
        count+=1 
    if count >= 3 :
      boolVar = False
      for idx , e in enumerate(tokens):
        if (idx == 0):
          continue
        else :
          if(e == redun )and(boolVar==False):
              token[idx] = 1
              boolVar=True
          else : 
              token[idx] = 0
  return tokens
```

```python
with open('/content/drive/MyDrive/Final folder/pwnWords.txt', 'r', encoding="utf-8") as f:
    words = f.readlines()
    words = [word.strip() for word in words]

with open('/content/drive/MyDrive/Final folder/pwnTransModel.txt','w') as f1 :
  for i,word in enumerate(words):
          input_ids = encode_input_str(
              text = word+'.',
              target_lang = 'ar',
              tokenizer = tokenizer,
              seq_len = 64,
              lang_token_map = LANG_TOKEN_MAPPING)
          input_ids = input_ids.unsqueeze(0).cuda()
          output_tokens = model.generate(input_ids, num_beams=10, num_return_sequences=1)
          translation = tokenizer.decode(output_tokens[0], skip_special_tokens=True)
          print(i)
          f1.write(translation)
          f1.write('\n')
```

```python
test_word = 'During the breathtaking sunset, a group of friends gathered on the sandy beach to enjoy a delicious picnic.'
input_ids = encode_input_str(
              text = test_word,
              target_lang = 'ar',
              tokenizer = tokenizer,
              seq_len = 100,
              lang_token_map = LANG_TOKEN_MAPPING)
input_ids = input_ids.unsqueeze(0).cuda()
print('Truncated input text:', tokenizer.convert_tokens_to_string(
    tokenizer.convert_ids_to_tokens(input_ids[0])))
output_tokens = model.generate(input_ids, num_beams=10,max_new_tokens = 1000 ,num_return_sequences=1)
for token_set in output_tokens:
  print(tokenizer.decode(token_set, skip_special_tokens=True))
```

```python
import nltk
from nltk.corpus import wordnet as wn
nltk.download('wordnet')

synsets = wn.all_synsets()


with open('/content/drive/MyDrive/Final folder/pwnExamplesWirhModel.txt','w') as f1 :
  for i,synset in enumerate(synsets):
    examples =synset.examples()
    if(examples):
          input_ids = encode_input_str(
              text = examples[0],
              target_lang = 'ar',
              tokenizer = tokenizer,
              seq_len = max_seq_len,
              lang_token_map = LANG_TOKEN_MAPPING)
          input_ids = input_ids.unsqueeze(0).cuda()
          output_tokens = model.generate(input_ids, num_beams=10, num_return_sequences=1)
          translation = tokenizer.decode(output_tokens[0], skip_special_tokens=True)
          print(i)
          f1.write(translation)
          f1.write('\n')
    else:
      f1.write('None')
      f1.write('\n')
```
