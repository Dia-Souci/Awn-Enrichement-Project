```python
from transformers import BertTokenizer, BertModel

model_name = 'bert-base-uncased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)
```

```python
sentence = "separated at the joint"
word = "مما أدى إلى التعليم "

tokens = tokenizer.tokenize(sentence)
word_tokens = tokenizer.tokenize(word)
```

```python
import torch

token_ids = tokenizer.convert_tokens_to_ids(tokens)
word_token_ids = tokenizer.convert_tokens_to_ids(word_tokens)

with torch.no_grad():
    inputs = torch.tensor([token_ids])
    outputs = model(inputs)
    sentence_embeddings = outputs.last_hidden_state.mean(dim=1).squeeze()

    word_inputs = torch.tensor([word_token_ids])
    word_outputs = model(word_inputs)
    word_embedding = word_outputs.last_hidden_state.mean(dim=1).squeeze()
```

```python
import torch.nn.functional as F

# Calculate cosine similarity
cosine_sim = F.cosine_similarity(sentence_embeddings, word_embedding, dim=0).item()

print(f"Similarity between the sentence and word: {cosine_sim}")
```

```python
def verrify_example(example,definition , translator = trans , model = model , similarity_fn = F.cosine_similarity):
  tokens = tokenizer.tokenize(example)
  word_tokens = tokenizer.tokenize(definition)
  token_ids = tokenizer.convert_tokens_to_ids(tokens)
  word_token_ids = tokenizer.convert_tokens_to_ids(word_tokens)

  with torch.no_grad():
      inputs = torch.tensor([token_ids])
      outputs = model(inputs)
      sentence_embeddings = outputs.last_hidden_state.mean(dim=1).squeeze()

      word_inputs = torch.tensor([word_token_ids])
      word_outputs = model(word_inputs)
      word_embedding = word_outputs.last_hidden_state.mean(dim=1).squeeze()
  cosine_sim = similarity_fn(sentence_embeddings, word_embedding, dim=0).item()
  return cosine_sim
```

```python
def sentenceSimilarityWithBert(model,tokenizer,definition , translated) :
  tokens = tokenizer.tokenize(definition)
  word_tokens = tokenizer.tokenize(translated)

  token_ids = tokenizer.convert_tokens_to_ids(tokens)
  word_token_ids = tokenizer.convert_tokens_to_ids(word_tokens)

  with torch.no_grad():
      inputs = torch.tensor([token_ids])
      outputs = model(inputs)
      sentence_embeddings = outputs.last_hidden_state.mean(dim=1).squeeze()

      word_inputs = torch.tensor([word_token_ids])
      word_outputs = model(word_inputs)
      word_embedding = word_outputs.last_hidden_state.mean(dim=1).squeeze()
  cosine_sim = F.cosine_similarity(sentence_embeddings, word_embedding, dim=0).item()

  return cosine_sim

def back_forth(definition,arabicdef,model=model,tokenizer=tokenizer,similarity_fn =sentenceSimilarityWithBert , translator = trans ):
  englishdef = translator.translate(arabicdef, dest='en').text
  sim = similarity_fn(model,tokenizer,definition , englishdef)
  return sim
```

```python
translated_model=[]
from transformers import AutoTokenizer , AutoModelForSeq2SeqLM
model_repo = 'google/mt5-small'
model_path = '/content/drive/MyDrive/MT5/v2MT5_AR_EN.pt'
max_seq_len = 20
```

```python
LANG_TOKEN_MAPPING = {
    'en': '<en>',
    'ar': '<ar>'
}
```

```python
special_tokens_dict = {'additional_special_tokens': list(LANG_TOKEN_MAPPING.values())}
tokenizer_ = AutoTokenizer.from_pretrained(model_repo)
tokenizer_.add_special_tokens(special_tokens_dict)
model = AutoModelForSeq2SeqLM.from_pretrained(model_repo)
model.resize_token_embeddings(len(tokenizer_))
```

```python
model.cuda()
model.load_state_dict(torch.load(model_path))
```

```python
def encode_input_str(text, target_lang, tokenizer, seq_len,
                     lang_token_map=LANG_TOKEN_MAPPING):
  target_lang_token = lang_token_map[target_lang]

  # Tokenize and add special tokens
  input_ids = tokenizer.encode(
      text = target_lang_token + text,
      return_tensors = 'pt',
      padding = 'max_length',
      truncation = True,
      max_length = seq_len)

  return input_ids[0]

test_sentence ='relational'
print('Raw input text:', test_sentence)

input_ids = encode_input_str(
    text = test_sentence,
    target_lang = 'ar',
    tokenizer = tokenizer,
    seq_len = 64,
    lang_token_map = LANG_TOKEN_MAPPING)
input_ids = input_ids.unsqueeze(0).cuda()

print('Truncated input text:', tokenizer.convert_tokens_to_string(
    tokenizer.convert_ids_to_tokens(input_ids[0])))


output_tokens = model.generate(input_ids, num_beams=10, num_return_sequences=1)
strr=''
# print(output_tokens)
for token_set in output_tokens:
  strr += tokenizer.decode(token_set, skip_special_tokens=True)
print(strr)
```

```python
def translate_reverso(example,ReversoApi = ReversoContextAPI):
  API = ReversoApi(example , '' , "en","ar")
  results = API.get_translations()
  translations = list(results)
  if(translations):
    translation = translations[0].translation
  else:
    translation = 'None'
  return translation
```
