```python
from transformers import BertTokenizer, BertModel

model_name = 'bert-base-uncased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)
```

```python
sentence = "separated at the joint"
word = "مما أدى إلى التعليم "

tokens = tokenizer.tokenize(sentence)
word_tokens = tokenizer.tokenize(word)
```

```python
import torch

token_ids = tokenizer.convert_tokens_to_ids(tokens)
word_token_ids = tokenizer.convert_tokens_to_ids(word_tokens)

with torch.no_grad():
    inputs = torch.tensor([token_ids])
    outputs = model(inputs)
    sentence_embeddings = outputs.last_hidden_state.mean(dim=1).squeeze()

    word_inputs = torch.tensor([word_token_ids])
    word_outputs = model(word_inputs)
    word_embedding = word_outputs.last_hidden_state.mean(dim=1).squeeze()
```

```python
import torch.nn.functional as F

# Calculate cosine similarity
cosine_sim = F.cosine_similarity(sentence_embeddings, word_embedding, dim=0).item()

print(f"Similarity between the sentence and word: {cosine_sim}")
```
