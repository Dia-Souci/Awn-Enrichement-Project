```python
from transformers import AutoTokenizer, AutoModel
import torch
import torch.nn.functional as F

model_name = "aubmindlab/bert-base-arabertv2"
tokenizer = AutoTokenizer.from_pretrained("aubmindlab/bert-base-arabertv2")
model = AutoModel.from_pretrained(model_name)
model = model.cuda()

tokens = ["example", "sentence", "tokens"]
word_tokens = ["word", "tokens"]

token_ids = tokenizer.convert_tokens_to_ids(tokens)
word_token_ids = tokenizer.convert_tokens_to_ids(word_tokens)

with torch.no_grad():
    inputs = torch.tensor([token_ids]).cuda()
    outputs = model(inputs)
    sentence_embeddings = outputs.last_hidden_state.mean(dim=1).squeeze()

    word_inputs = torch.tensor([word_token_ids]).cuda()
    word_outputs = model(word_inputs)
    word_embedding = word_outputs.last_hidden_state.mean(dim=1).squeeze()

# Calculate cosine similarity
cosine_sim = F.cosine_similarity(sentence_embeddings, word_embedding, dim=0).item()


print(f"Similarity between the sentence and word: {cosine_sim}")
```

```python
def calculate_cosine_similarity_arabert(tokens, word_tokens , model , tokenizer):
    token_ids = tokenizer.convert_tokens_to_ids(tokens)
    word_token_ids = tokenizer.convert_tokens_to_ids(word_tokens)

    with torch.no_grad():
        inputs = torch.tensor([token_ids]).cuda()
        outputs = model(inputs)
        sentence_embeddings = outputs.last_hidden_state.mean(dim=1).squeeze()

        word_inputs = torch.tensor([word_token_ids]).cuda()
        word_outputs = model(word_inputs)
        word_embedding = word_outputs.last_hidden_state.mean(dim=1).squeeze()

    # Calculate cosine similarity
    cosine_sim = F.cosine_similarity(sentence_embeddings, word_embedding, dim=0).item()

    return cosine_sim
```

```python
from transformers import TFAutoModelForSequenceClassification
model_class = TFAutoModelForSequenceClassification.from_pretrained("/content/drive/MyDrive/ArabertModel/weights2")
```

```python
import tensorflow as tf

def tokenize_function_raw(example , prep = arabert_prep):
    text = prep.preprocess(example)
    return tokenizer(text)

def predict_label(sentence , model = model_class ):
    # Tokenize the sentence
    input = tokenize_function_raw(sentence)["input_ids"]

    # Make the prediction
    predictions = model.predict([input])
    predicted_label = tf.argmax(predictions.logits, axis=1).numpy()[0]

    return predicted_label
```

```python
sentence = 'دعونا الاستفادة من جميع الهيئات، وقادرة أو لا'
predicted_label = predict_label(sentence , model = model_class)
print("Input:", sentence)
print("Predicted label:", predicted_label)
```
